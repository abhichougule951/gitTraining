```python
# Separate human inputs and bot responses
human_inputs = [item[0] for item in data]
print("human_inputs :",human_inputs)
bot_responses = [item[1] for item in data]
print("bot_responses :",bot_responses)
# Tokenization
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(human_inputs + bot_responses)
print("tokenizer :",tokenizer)
# Convert text data to sequences
human_sequences = tokenizer.texts_to_sequences(human_inputs)
bot_sequences = tokenizer.texts_to_sequences(bot_responses)
print("human_sequences :",human_sequences)
print("bot_sequences :",bot_sequences)
# Pad sequences for a fixed input length
max_seq_length = max(max(len(seq) for seq in human_sequences), max(len(seq) for seq in bot_sequences))
human_padded = tf.keras.preprocessing.sequence.pad_sequences(human_sequences, maxlen=max_seq_length, padding='post')
bot_padded = tf.keras.preprocessing.sequence.pad_sequences(bot_sequences, maxlen=max_seq_length, padding='post')

# Encode output labels into binary format
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(bot_responses)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(human_padded, y_encoded, test_size=0.3, random_state=42)

# Build LSTM model
model = Sequential()
model.add(Embedding(len(tokenizer.word_index) + 1, 128, input_length=max_seq_length))
model.add(LSTM(35))
model.add(Dense(len(label_encoder.classes_), activation='softmax'))

# Compile model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])



```

```python
# Train model
model.fit(X_train, y_train, epochs=300, batch_size=12, validation_data=(X_test, y_test))
```
